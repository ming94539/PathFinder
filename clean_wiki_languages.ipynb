```python
file = open("wiki_languages.txt","r")
lang = file.readlines()
```

```python
clean = []
for l in lang:
    if "(" in l:
        p_i = l.index("(")
        l = l[:p_i]
    if " –" in l:
        p_i = l.index(" –")
        l = l[:p_i]
    clean.append(l.strip().lower())
clean = list(set(clean))
clean = sorted(clean)
```

```python
clean
```

```python
languages = ['python','c','c++','java','perl','go','dart','javascript','lua','lisp','html','css','scss',
            'c#','objective-c','r','.net','php','sql','nosql','matlab','pascal','swift','visual basic','cobol'
            ,'scala','abap','rust','fortran','prolog','ada','julia','haskell','kotlin','bash','scheme','postgresql'
            ,'mysql','sqlite','racket','typescript','ruby','verilog']
languages = sorted(languages)
#objective-c,
#lisp-ISO
for i in languages:
    if i not in clean:
        print(i)
```

```python
len(languages)
```

```python
ss = open("lists/final_keywords.txt",'r')
skills = ss.readlines()
print(len(skills))
skill_lang = []
for s in skills:
    if s.strip() in clean:
        skill_lang.append(s.strip())
print(len(skill_lang))
skill_lang = sorted(skill_lang)
not_in_lang = []
for g in skill_lang:
    if g not in languages:
        not_in_lang.append(g)
print(len(not_in_lang))
print(not_in_lang)
```

```python
s2 = open("final_keywords.txt",'r')
skills = s2.readlines()
print(len(skills))
new_skills = []
for s in skills:
    if s.strip() not in clean:
        new_skills.append(s.strip())
print(len(new_skills))
new_skills = sorted(new_skills)
new_skills
```

```python
with open('final_keywords2.txt', 'w') as filehandle:
    for listitem in new_skills:
        filehandle.write('%s\n' % listitem)
```
