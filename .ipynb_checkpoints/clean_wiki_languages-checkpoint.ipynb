```python
file = open("wiki_languages.txt","r")
lang = file.readlines()
```

```python
clean = []
for l in lang:
    if "(" in l:
        p_i = l.index("(")
        l = l[:p_i]
    clean.append(l.strip().lower())
clean = list(set(clean))
clean = sorted(clean)
```

```python
clean
```

```python
languages = ['python','c','c++','java','perl','go','dart','javascript','lua','lisp','html','c','c++'
            ,'c#','objective-c','r','.net','php','sql','nosql','matlab','pascal','swift']
#objective-c
#lisp-ISO
for i in languages:
    if i not in clean:
        print(i)
```
